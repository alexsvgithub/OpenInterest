```python
from bs4 import BeautifulSoup
import requests
import requests
import json
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
```

```python
new_url = 'https://www.nseindia.com/api/option-chain-indices?symbol=NIFTY'
headers = {'User-Agent': 'Mozilla/5.0'}
page = requests.get(new_url,headers=headers)
dajs = json.loads(page.text)
plotting = []



def fetch_oi(expiry_dt):
    
    ce=[]
    pe=[]
    ceOi=[]
    peOi=[]
    cceOi=[]           #changed open interest
    cpeOi=[]
    ce_values = [data['CE'] for data in dajs['records']['data'] if "CE" in data and data['expiryDate'] == expiry_dt]
    pe_values = [data['PE'] for data in dajs['records']['data'] if "PE" in data and data['expiryDate'] == expiry_dt]
    ce_dt = pd.DataFrame(ce_values).sort_values(['strikePrice'])
    pe_dt = pd.DataFrame(pe_values).sort_values(['strikePrice'])
    ce= [n for n in ce_dt['strikePrice'] if n % 100 == 0]
    for i in ce:
        ceOi.append(int(ce_dt.loc[ce_dt.strikePrice==i]['openInterest']))    
        cceOi.append(int(ce_dt.loc[ce_dt.strikePrice==i]['changeinOpenInterest']))
    pe= [n for n in pe_dt['strikePrice'] if n % 100 == 0]
    for i in pe:
        peOi.append(int(pe_dt.loc[pe_dt.strikePrice==i]['openInterest']))
        cpeOi.append(int(pe_dt.loc[pe_dt.strikePrice==i]['changeinOpenInterest']))
    return ce, ceOi,pe,peOi,cceOi, cpeOi;
    
   

def main():
    import matplotlib.pyplot as plt
    
    ce=[]
    pe=[]
    ceOi=[]
    peOi=[]
    cceOi=[]
    cpeOi=[]
    expiry_dt = '20-Jan-2022'
    ce,ceOi,pe,peOi,cceOi,cpeOi= fetch_oi(expiry_dt)
#     print(ce)
#     print(pe)
#     print(ceOi)
#     print(peOi)
#     print(cceOi)
#     print(cpeOi)
#     fetch_oi(expiry_dt)

   
    
    import pandas as pd
   
    plotdata = pd.DataFrame({

        "Call":ceOi,     #calls

        "Put":peOi},   #puts

        index=[ce,pe]) #strikes

    plotdata.plot(kind="bar",figsize=(15, 5),color=['g','r'])
    
#     plt.axvline(x = 18300, color = 'b', label = 'axvline - full height')
    

    plt.title("Open Interest")

    plt.xlabel("Strike Price")

    plt.ylabel("Open Interest")
    
    
    ##################################################Changed Open Interest
    
    
    plotdata = pd.DataFrame({

        "Call":cceOi,     #calls

        "Put":cpeOi},   #puts

        index=[ce,pe]) #strikes

    plotdata.plot(kind="bar",figsize=(15, 3),color=['g','r'])
    
#     plt.axvline(x = 18300, color = 'b', label = 'axvline - full height')
    

    plt.title("Change in Open Interest")

    plt.xlabel("Strike Price")

    plt.ylabel("Open Interest")
    


if __name__ == '__main__':
    main()
```
